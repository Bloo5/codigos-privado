```python
#List

list_class = ['Warrior', 'Mage', 'Rogue', 'Ranger', 'Priest']

print('first:  ', list_class[0]   )
print('last:   ', list_class[-1]  )
print('range:  ', list_class[2:5] )
print('begging:', list_class[:3]  )
print('end:    ', list_class[3:]  )

#Fiding in List
if "Mage" in list_class:
   print('Existe')
else:
   print('Não Existe')

#Changing List
list_class[1] = 'Sorcerer'

#Chaging from index 2 to 5
list_class[2:5] = ['Assassin', 'Bard', 'Paladin']

#Insert in List
list_class.insert(2, 'Warlock')

#Insert in the last index
list_class.append('Monk')

#Extend List
list_class_2 = ['Rogue', 'Ranger', 'Mage']
list_class.extend(list_class_2)

#Remove Item
list_class.remove('Assassin')
list_class.pop(2)
list_class.pop()   #remove ultimo item

print(list_class)
```

```python
#Class

class Character:
   def __init__(self, Name, Race, Class ,Age, Gender, Main):
      self.Name = Name
      self.Race = Race
      self.Class = Class
      self.Age = Age
      self.Gender = Gender
      self.Main = Main

   def sayName(self, ):
      return('My Name is '+ self.Name)

Ocha = Character('Ocha','Elf', 'Mage', 27, 'Female', False)

print(Ocha.Name)
print(Ocha.Race)
print(Ocha.Class)
print(Ocha.Age)
print(Ocha.Gender)
print(Ocha.Main)
print(Ocha.sayName())
```

```python
#Inheritance

class Character:
  def __init__(self, Name, Race, Age, Gender, Main):
    self.Name = Name
    self.Race = Race
    self.Age = Age
    self.Gender = Gender
    self.Main = Main
    self.Level = 1
#####################################################################################
class Mage(Character):
  def __init__(self, Name, Race ,Age, Gender, Main):
    super().__init__( Name, Race ,Age, Gender, Main )
    self.Class= 'Mage'
    self.Mana = 100
    self.Health = 80
    self.Barrier = 0

  def FireBall(self, target):
    self.Mana = self.Mana - 10
    target.Health = target.Health - 20
  
  def IceBarrier(self):
    self.Barrier = self.Barrier + 20
    self.Mana = self.Mana - 10

  def Meditate(self):
    self.Mana = self.Mana + 20
#####################################################################################
class Warrior(Character):
  def __init__(self, Name, Race ,Age, Gender, Main):
    super().__init__( Name, Race ,Age, Gender, Main )
    self.Class= 'Warrior'
    self.Rage = 0
    self.Health = 150
    self.Barrier = 50

  def Decimate(self, target):
    target.Health = target.Health - 5 + self.Rage
    self.Rage = 0
  
  def ShieldUp(self):
    self.Barrier = self.Barrier + 10

  def AxeSlash(self, target):
    target.Health = target.Health - 5
#####################################################################################
class Rogue(Character):
  def __init__(self, Name, Race, Age, Gender, Main):
    super().__init__( Name, Race, Age, Gender, Main )
    self.Class= 'Rogue'
    self.Mana = 50
    self.Health = 110
    self.Barrier = 0
    self.Sneaky = False

  def SneakyAtack(self, target):
    if(self.Sneaky == True):
      self.Mana = self.Mana - 10
      target.Health = target.Health - 40
      self.Sneaky = False
    else:
      self.Mana = self.Mana - 10
      target.Health = target.Health - 10

  def ShadowCloak(self):
    self.Mana = self.Mana - 5
    self.Sneaky = True

  def CheapShot(self, target):
    self.Mana = self.Mana + 10
    target.Health = target.Health - 5
#####################################################################################
class Priest(Character):
  def __init__(self, Name, Race ,Age, Gender, Main):
    super().__init__( Name, Race ,Age, Gender, Main )
    self.Class= 'Priest'
    self.Mana = 100
    self.Health = 80
  
  def Holy(self, target):
    if(target):
      target.Health = target.Health + 15
    else:
      self.Health = self.Health + 20
    self.Mana = self.Mana - 20

  def DrainMana(self, target):
    if(target.Mana):
      target.Mana = target.Mana - 10
      self.Mana = self.Mana + 10
#####################################################################################
Ocha  = Mage ('Ocha' , 'Elf'  , 27, 'Female', False)
Makui = Rogue('Makui', 'Human', 32, 'Male'  , False)

Ocha.FireBall(Makui)
print(Makui.Health)
```

```python
#Conversôes
decimal = 100
#Converter para binario
def to_binary(n):
  return bin(n)

#Converter para Hexadecimal
def to_hex(dec):
  return hex(dec)

print(to_binary(decimal), 'Binario')
print(to_hex(decimal), 'Hexadecimal')
```
