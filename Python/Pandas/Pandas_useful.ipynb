```python
import pandas as pd

def split_dataframe(df, fraction=0.5):
    if not (0 <= fraction <= 1):
        raise ValueError("Fraction must be between 0 and 1.")
    
    split_index = int(len(df) * fraction)
    
    df1 = df.iloc[:split_index].reset_index(drop=True)
    df2 = df.iloc[split_index:].reset_index(drop=True)
    
    return df1, df2

def split_dataframe_random(df, fraction=0.5):
    if not (0 <= fraction <= 1):
        raise ValueError("Fraction must be between 0 and 1.")
    
    df_shuffled = df.sample(frac=1).reset_index(drop=True)
    
    split_index = int(len(df_shuffled) * fraction)
    
    df1 = df_shuffled.iloc[:split_index].reset_index(drop=True)
    df2 = df_shuffled.iloc[split_index:].reset_index(drop=True)
    
    return df1, df2
```
